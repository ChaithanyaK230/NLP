```python
corpus=""" Hello, My name is Chaithanya's  . 
I am learning Natural language processing . 
I want to all the things i can do with it and learn everything."""
print(corpus)
```

###Tokenization --> Sentence to Paragraph

```python
import nltk
print(nltk.__version__)  # just to confirm

# Download required models
nltk.download("punkt")
nltk.download("punkt_tab")  # important for new NLTK versions

```

```python
from nltk.tokenize import sent_tokenize
print(sent_tokenize(corpus))
```

```python
documents=sent_tokenize(corpus)
print(documents)
print(type(documents))
for sentence in documents:
    print(sentence)
```

###Tokenization :
1.Paragragh to words.
2.Sentences to words.

```python
from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
print(words)
for word in words:
    print(word)
print(type(word))
print(type(words))
```

```python
documents=sent_tokenize(corpus)
for i in documents:
    print(word_tokenize(i))
```

### TO Tokenize in accurate way beacsue 's is not handled properly in the above approach and every punctuation is treated as a separate word  so to handle it properly we can use wordpunct_tokenize

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
TreebankWordTokenizer().tokenize(corpus)
```
