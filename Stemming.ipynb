```python
words=["eating","eats","learns","learning","goes","going","played","playing","plays","finally","final"]
```

```python
from nltk.stem import PorterStemmer
```


Major disadvantages of stemming is some words does not get converted to its stem word and it changes the caomplete meaning . Eg: Goes---> Goe{which changes the meaning.}

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(f"{word}---->{stemming.stem(word)}")
```

```python
stemming.stem("Congratulations")
```

```python
from nltk.stem import RegexpStemmer
reg_stemmer=RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem("ingeating")
```

Better Technique


```python
from nltk.stem import SnowballStemmer
snow_stemmer=SnowballStemmer("english")
```

```python
snow_stemmer.stem("goes")
```

```python
print(stemming.stem('fairly'))
print(snow_stemmer.stem('fairly'))
```
