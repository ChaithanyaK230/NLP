```python
corpus="""Artificial intelligence (AI) refers to computer systems that can perform tasks usually associated with human thinking, such as recognizing speech, understanding language, or making decisions. Modern AI systems learn patterns from large amounts of data instead of being explicitly programmed for every rule, which allows them to improve at tasks like image recognition, translation, and recommendation over time. Today, AI is embedded in everyday tools—search engines, navigation apps, chatbots, and recommendation systems—often working quietly in the background to personalize experiences and automate routine work.​At the same time, AI raises important questions about fairness, privacy, and the future of work, because these systems can influence hiring, lending, and access to information. Researchers and policymakers are working on guidelines for responsible and ethical AI, focusing on transparency, reducing bias in models, and ensuring humans remain in control of critical decisions. As AI capabilities continue to grow—especially in areas like generative models that can create text, images, and video—society must balance innovation with safeguards so that these technologies are used to support human well‑being rather than replace or harm it."""
```

```python
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(corpus)
sentences
```

```python
import nltk

nltk.download("averaged_perceptron_tagger", download_dir=r"D:\nltk_data")
nltk.download("averaged_perceptron_tagger_eng", download_dir=r"D:\nltk_data")

nltk.data.path.append(r"D:\nltk_data")
```

```python
all_words=[]
for i in range(len(sentences)):
    word=nltk.word_tokenize(sentences[i])
    pos_tags=nltk.pos_tag(word)
    all_words.append(pos_tags)
    print(pos_tags)

sum=0
for item in all_words:
    sum=sum+len(item)
print(sum)
```

```python
from nltk.corpus import stopwords
```

```python
all_words2=[]
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[word for word in words if word not in set(stopwords.words("english"))]
    pos_tags=nltk.pos_tag(words)
    all_words2.append(pos_tags)
    print(pos_tags)

sum=0
for item in all_words2:
    sum=sum+len(item)
print(sum)
```
